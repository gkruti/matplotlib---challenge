## Observations and Insights 

```python
# Dependencies and Setup

import matplotlib.pyplot as plt
from scipy.stats import linregress
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files

mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

mouse_metadata.head()
study_results.head()

mousedata_df = pd.merge(mouse_metadata, study_results, on = "Mouse ID")

mousedata_df.head()

# Combine the data into a single dataset

# Display the data table for preview
```

```python
# Checking the number of mice.

mousedata_df["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_mice = mousedata_df.loc[mousedata_df.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"]

duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

clean_df = mousedata_df[mousedata_df["Mouse ID"].isin(duplicate_mice)==False]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.

clean_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

regimen = clean_df.groupby(["Drug Regimen"])

mean_regimen = regimen["Tumor Volume (mm3)"].mean()

median_regimen = regimen["Tumor Volume (mm3)"].median()

var_regimen = regimen["Tumor Volume (mm3)"].var()

std_regimen = regimen["Tumor Volume (mm3)"].std()

sem_regimen = regimen["Tumor Volume (mm3)"].sem()

regimen_df = pd.DataFrame({"Mean": mean_regimen, "Median": median_regimen, "Variance": var_regimen,
"Standard Deviation": std_regimen, "SEM": sem_regimen})

regimen_df

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function

single_groupby = clean_df.groupby('Drug Regimen')
table_stats = single_groupby.agg(['mean','median','var','std','sem'])["Tumor Volume (mm3)"]
table_stats
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen


correl=round(st.pearsonr(merge_cap['Weight (g)'],merge_cap['Tumor Volume (mm3)'])[0],2)

x_values = merge_cap['Weight (g)']
y_values = merge_cap['Tumor Volume (mm3)']

print("The correlation between mouse weight and average tumor volume is " + str(correl))


```

```python

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Mouse Weight')
plt.ylabel('Avg Tumor Volume')
plt.show()
```
